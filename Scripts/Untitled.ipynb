```python
from sklearn.model_selection import KFold
import pandas as pd
```

```python
def kFold(inputs):
        kfold = KFold(5, True, 11)
        idxs = []

        for train, test in enumerate(kfold.split(inputs)):
            idxs.append((train, test))

        return idxs
```

```python
csv_data = pd.read_csv('../data_csv/0_9_all.csv')
```

```python
csv_list = []
for i in csv_data.itertuples():
    csv_list += [i[1] + ',' + str(i[2])]
len(csv_list)
```

```python
idxs = kFold(csv_list[0:5400])
```
